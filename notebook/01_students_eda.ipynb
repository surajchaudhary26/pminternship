```python

import pandas as pd

# Paths
raw_path = "../ml_engine/data/raw/students_.csv"
noisy_path = "../ml_engine/data/processed/students_noisy.csv"
random_path = "../ml_engine/data/processed/students_noisy_random.csv"
balanced_path = "../ml_engine/data/processed/students_balanced.csv"

# Load datasets
students_raw = pd.read_csv(raw_path)            # Original synthetic clean dataset
students_noisy = pd.read_csv(noisy_path)        # Augmented noisy dataset (typos, phone noise, etc.)
students_random = pd.read_csv(random_path)      # Randomized missing values dataset (realistic messy)
students_balanced = pd.read_csv(balanced_path)  # Cleaned & skill-rebalanced dataset (ML-ready)

# Show shapes
print("üìÇ Dataset Shapes:")
print("Raw (synthetic):", students_raw.shape)
print("Noisy (augment.py):", students_noisy.shape)
print("Random Missing (augment_random_missing.py):", students_random.shape)
print("Balanced (augment_skills.py):", students_balanced.shape)

# Preview first 3 rows of each
print("\nüîç Raw (synthetic sample):")
display(students_raw.head(3))

print("\nüîç Noisy (augment.py sample):")
display(students_noisy.head(3))

print("\nüîç Random Missing (augment_random_missing.py sample):")
display(students_random.head(3))

print("\nüîç Balanced (augment_skills.py sample):")
display(students_balanced.head(3))
```

```python
datasets = {
    "Raw": students_raw,
    "Noisy": students_noisy,
    "Random": students_random,
    "Balanced": students_balanced
}

missing_summary = {}
for name, df in datasets.items():
    missing_summary[name] = df.isnull().sum()

missing_df = pd.DataFrame(missing_summary)
print("üîç Missing Values Comparison Across Versions")
display(missing_df)
```

```python
from collections import Counter
import ast
def extract_skills(df):
    skills_list = []
    for row in df["Skills"].dropna():
        try:
            skills = ast.literal_eval(row) if isinstance(row, str) else row
            skills_list.extend(skills)
        except:
            continue
    return Counter(skills_list)

skill_counts_raw = extract_skills(students_raw).most_common(10)
skill_counts_noisy = extract_skills(students_noisy).most_common(10)
skill_counts_random = extract_skills(students_random).most_common(10)
skill_counts_balanced = extract_skills(students_balanced).most_common(10)

print("üîç Top Skills (Raw):", skill_counts_raw)
print("üîç Top Skills (Noisy):", skill_counts_noisy)
print("üîç Top Skills (Random):", skill_counts_random)
print("üîç Top Skills (Balanced):", skill_counts_balanced)
```

```python
# Load v2 Dataset

students_v2 = pd.read_csv("../ml_engine/data/processed/students_balanced_v2.csv")

print("Shape:", students_v2.shape)
students_v2.head()
```

```python
# Top 15 Skills (v2)
from collections import Counter
import ast
import matplotlib.pyplot as plt
import seaborn as sns

skills_list = []

for row in students_v2["Skills"].dropna():
    try:
        skills = ast.literal_eval(row) if isinstance(row, str) else row
        skills_list.extend(skills)
    except:
        continue

skill_counts = Counter(skills_list).most_common(15)

skills_df = pd.DataFrame(skill_counts, columns=["Skill", "Count"])
print("üîç Top 15 Skills (Balanced v2):\n", skills_df)

plt.figure(figsize=(10,5))
sns.barplot(data=skills_df, x="Skill", y="Count", palette="rocket")
plt.title("Top 15 Skills in Students Dataset (Balanced v2)")
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))
sns.countplot(x=students_v2["Preferred_Job_Type"], palette="pastel")
plt.title("Preferred Job Type (Balanced v2 Dataset)")
plt.xlabel("Job Type")
plt.ylabel("Count")
plt.show()
```

```python
# Count job types
job_type_counts = students_v2["Preferred_Job_Type"].value_counts()

print("üîç Preferred Job Type Counts (Balanced v2):")
print(job_type_counts)

# Plot distribution
plt.figure(figsize=(6,4))
sns.countplot(x=students_v2["Preferred_Job_Type"], palette="pastel",
              order=job_type_counts.index)  # order by counts
plt.title("Preferred Job Type (Balanced v2 Dataset)")
plt.xlabel("Job Type")
plt.ylabel("Count")
plt.show()
```
