```python
import pandas as pd

internships_raw = pd.read_csv("../ml_engine/data/raw/internships_.csv")

print("Shape:", internships_raw.shape)
internships_raw.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

missing = internships_raw.isnull().sum().sort_values(ascending=False)

print("🔍 Missing values per column:")
print(missing)

plt.figure(figsize=(10,5))
sns.barplot(x=missing.index, y=missing.values, palette="viridis")
plt.xticks(rotation=45)
plt.title("Missing Values in Internships Dataset (Raw)")
plt.ylabel("Count of Missing Values")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(internships_raw["Stipend"], bins=30, kde=True, color="green")
plt.title("Stipend Distribution (Raw Internships Dataset)")
plt.xlabel("Stipend (INR)")
plt.ylabel("Number of Internships")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(y=internships_raw["Duration"], order=internships_raw["Duration"].value_counts().index, palette="pastel")
plt.title("Duration Distribution (Raw Internships Dataset)")
plt.xlabel("Count")
plt.ylabel("Duration")
plt.show()
```

```python
top_titles = internships_raw["Job_Title"].value_counts().head(10)
print("🔍 Top 10 Job Titles (Raw):")
print(top_titles)

plt.figure(figsize=(8,5))
sns.barplot(x=top_titles.values, y=top_titles.index, palette="rocket")
plt.title("Top 10 Internship Job Titles")
plt.xlabel("Count")
plt.ylabel("Job Title")
plt.show()
```

```python
print("Columns in internships_raw dataset:")
print(internships_raw.columns.tolist())
```

```python
from collections import Counter
import ast

skills_list = []

for row in internships_raw["Skills_Required"].dropna():
    try:
        skills = ast.literal_eval(row) if isinstance(row, str) else row
        skills_list.extend(skills)
    except:
        continue

skill_counts = Counter(skills_list).most_common(15)

skills_df = pd.DataFrame(skill_counts, columns=["Skill", "Count"])
print("🔍 Top 15 Skills (Raw Internships):\n", skills_df)

plt.figure(figsize=(10,5))
sns.barplot(data=skills_df, x="Skill", y="Count", palette="rocket")
plt.title("Top 15 Skills in Internships Dataset (Raw)")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd

internships_v2 = pd.read_csv("../ml_engine/data/processed/internships_balanced_v2.csv")

print("Shape:", internships_v2.shape)
internships_v2.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

missing = internships_v2.isnull().sum().sort_values(ascending=False)

print("🔍 Missing values per column:")
print(missing)

plt.figure(figsize=(10,5))
sns.barplot(x=missing.index, y=missing.values, palette="viridis")
plt.xticks(rotation=45)
plt.title("Missing Values in Internships Dataset (Balanced v2)")
plt.ylabel("Count of Missing Values")
plt.show()
```
