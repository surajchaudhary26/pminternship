```python
import pandas as pd

# Load matches file (already generated by pipeline)
matches = pd.read_csv("../ml_engine/data/processed/matches.csv")
pd.set_option("display.max_colwidth", None)

# üîé Function to get recommendations for a student
def get_recommendations(student_id: str, top_n: int = 5):
    row = matches[matches["student_id"] == student_id]
    if row.empty:
        print(f"‚ö†Ô∏è No recommendations found for {student_id}")
        return
    recs = eval(row.iloc[0]["top_matches"])
    print(f"\nüéì Recommendations for Student {student_id}:\n")
    for job_id, score in recs[:top_n]:
        print(f"  - Internship {job_id} (score: {score:.2f})")

# Example usage:
get_recommendations("S00033", top_n=3)
get_recommendations("S00032", top_n=3)
```

```python
import pandas as pd
from pathlib import Path

# Path setup (adjust if notebook folder is different)
proc = Path("../ml_engine/data/processed")
gap_file = proc / "matches_with_gaps.csv"
basic_file = proc / "matches.csv"

# Load gap-analysis recommendations if available
if gap_file.exists():
    matches = pd.read_csv(gap_file)
    print("‚úÖ Loaded gap-analysis recommendations")
else:
    matches = pd.read_csv(basic_file)
    print("‚ö†Ô∏è Gap-analysis file not found, loaded basic recommendations")

pd.set_option("display.max_colwidth", None)

# üîé Function to get recommendations for a student
def get_recommendations(student_id: str, top_n: int = 5):
    row = matches[matches["student_id"] == student_id].head(top_n)
    if row.empty:
        print(f"‚ö†Ô∏è No recommendations found for {student_id}")
        return
    
    print(f"\nüéì Recommendations for Student {student_id}:\n")
    for _, r in row.iterrows():
        job_id = r["job_id"]
        score = r["score"]
        missing = r.get("missing_skills", "")
        missing_pct = r.get("missing_pct", 0)
        print(f"  - Internship {job_id} (score: {score:.2f})")
        if "missing_skills" in matches.columns:
            print(f"      Missing skills: {missing} ({missing_pct}%)")

# Example usage:
get_recommendations("S00033", top_n=3)
get_recommendations("S00032", top_n=3)
```
